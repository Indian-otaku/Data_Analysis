```python
tree_data.describe()
```

The tree diameter is given in inches as mentioned in the data. But a tree having 450 inches diameter is not at all normal. Let's check whether it is an outlier or not.

But first of all to get a more accurate distribution of tree diameter, I only want to take into consideration trees that are dead or alive and not the stumps.


```python
dead_or_alive_trees = tree_data[
    (tree_data.status == "Alive") + (tree_data.status == "Dead")
]

dead_or_alive_trees["tree_dbh"].describe()
```

The above data still shows the possibility of outliers in the data. Lets check it using graphs.


```python
plt.figure(figsize=(12, 5))
sns.histplot(data=dead_or_alive_trees, x="tree_dbh")
plt.show()
```

They are indeed outliers. The actual distribution looks like exponential decay that ends before 100 inches tree diameter. Let's try plotting observations that are shown to have diameter above 100 inches.


```python
plt.figure(figsize=(12, 10))
sns.scatterplot(
    data=dead_or_alive_trees[dead_or_alive_trees.tree_dbh >= 100],
    x="tree_dbh",
    y="spc_common",
)
plt.show()
```

One plausible reason for why there is this large difference is probably due to change in measuring system used (its probably in cm rather than inch) or another plausible reasoning is that they forgot to divide by pi (3.14) as mentioned in the data description.

Lets check if stump diameter suffer from the same problem or not.


```python
tree_stump_data = tree_data[tree_data.status == "Stump"]

tree_stump_data.describe()
```

It does seem to be the case that stump diameter also has outliers. Let's visualize it.


```python
plt.figure(figsize=(12, 5))
sns.histplot(data=tree_stump_data, x="stump_diam")
plt.show()
```

Compared to tree diameter, the severity of outliers in stump diameter is very low.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x = np.array([2, 4, 6, 3])
y = np.array([5, 2, 5, 1])
z = np.array([7, 2, 3, 2])
```

```python
plt.bar(range(4), x, label="a")
plt.bar(range(4), y, bottom=x, label="b")
plt.bar(range(4), z, bottom=x + y, label="c")
plt.legend()
plt.xticks(range(4), ["company 1", "company 2",
           "company 3", "company 4"], rotation=0)
plt.xlabel("Company names")
plt.ylabel("Cumulative Expenses")
sns.despine(top=True, right=False, left=True, bottom=True, offset=True)
plt.show()
```
